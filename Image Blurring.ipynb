```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('flower.jpg')
```

```python
blur = cv.blur(img,(5,5),0)
```

```python
plt.subplot(121)   #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.imshow(img)
plt.title('Original')
plt.xticks([]) 
plt.yticks([])
plt.subplot(122)    #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.imshow(blur)
plt.title('Blurred')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
edges = cv.Canny(blur,100,200)
```

```python
plt.subplot(121)
plt.imshow(img,cmap = 'gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
contours, hierarchy = cv.findContours(edges,cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
# draw all contours
image = cv.drawContours(edges, contours, -1, (0, 255, 0), 2)
```
